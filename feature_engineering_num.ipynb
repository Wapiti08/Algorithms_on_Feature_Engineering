The notebook to explore different processing (scaling) order to the final result, I will discuss 
the following options:

For Deep Learning (Regression Problem only):

    - Scaling whole datasets for Deep Learning model
    - Scaling separate training and testing datasets for Deep Learning model

For Machine Learning (Regression Problem only):
    
    - Scaling whole datasets for Machine Learning model
    - Scaling separate training and testing datasets for Machine Learning model

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
# diaplay all the columns
pd.pandas.set_option("display.max_columns", None)

data = pd.read_csv("train.csv")

data.drop("Id", axis=1, inplace=True)
```

```python
# get the categorical and numerical featuers
car_feas = [fea for fea in data.columns if data[fea].dtype == 'O']
## add MSSubClass to cate features
car_feas = car_feas + ['MSSubClass']
num_feas = [fea for fea in data.columns if fea not in car_feas and fea not in ["SalePrice"]]
# get and test on numeric data only
num_data = data[num_feas]
```

```python
num_data
```

1. Scaling First for ML Model

```python
# processing data
for fea in num_feas:
    mean_val = data[fea].mean()
    # add binary missing indicator
    num_data[fea+'_na'] = np.where(num_data[fea].isnull(), 1, 0)
    # replace missing values by the mean
    num_data[fea].fillna(mean_val, inplace=True)
```

```python
num_data
```

```python
# scaling the training dataset first
scaler = MinMaxScaler()
sca_data = scaler.fit_transform(num_data)

# split the data
X_train, X_test, y_train, y_test = train_test_split(
                sca_data, 
                data['SalePrice'],
                test_size = 0.1,
                random_state=0)
```

```python
# train the model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Lasso

lin_model = Lasso(alpha=0.001, random_state=0)
# train the model
lin_model.fit(X_train, y_train)

# make predictions for train set
Y_train_pred = lin_model.predict(X_train)

# determine mse, rmse and r2
print('train mse: {}'.format(int(
    mean_squared_error(y_train, Y_train_pred))))
print('train rmse: {}'.format(int(
    mean_squared_error(y_train, Y_train_pred, squared=False))))
print('train r2: {}'.format(
    r2_score(y_train, Y_train_pred)))
print()


# make predictions for test set
Y_test_pred = lin_model.predict(X_test)

# determine mse, rmse and r2
print('test mse: {}'.format(int(
    mean_squared_error(y_test, Y_test_pred))))
print('test rmse: {}'.format(int(
    mean_squared_error(y_test, Y_test_pred, squared=False))))
print('test r2: {}'.format(
    r2_score(y_test, Y_test_pred)))
print()
```

3. Scaling First for DL Model

```python
# define all the necessary imports
from keras.models import Sequential
from keras.layers import Dense
```
